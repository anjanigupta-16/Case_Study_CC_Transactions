# Data Exploration and Cleaning
* Load the dataset into a Pandas DataFrame and display the first 5 rows.
* Check the shape, column names, and summary statistics of the dataset.
* Identify and handle missing values (fill or drop based on the data type).
* Convert Transaction_Date into datetime format and extract year, month, and day as new columns.

```python
#Load the dataset into a Pandas DataFrame and display the first 5 rows.
import pandas as pd

df = pd.read_csv("credit_card_transactions.csv")
df.head()
```

```python
#Check the shape, column names, and summary statistics of the dataset.
print(df.shape) #(raw,column)
print(df.columns.to_list())
print('summary statistics \n', df.describe()) #you can use include=all (for all details)
```

```python
#Identify and handle missing values (fill or drop based on the data type)
print(df.isnull().sum())
```

```python
#fill the missing value

df['Merchant']= df['Merchant'].fillna('Unknown')
df['Transaction_Type']= df['Transaction_Type'].fillna('Unknown')
df['Category']= df['Category'].fillna('Unknown')
df['Payment_Mode']= df['Payment_Mode'].fillna('Unknown')
df['Transaction_Status']= df['Transaction_Status'].fillna('Unknown')

#fill int value
df['Amount'] = df['Amount'].fillna(df['Amount'].median()) #you can also you mean() 
print(df.isnull().sum())
```

```python
#Convert Transaction_Date into datetime format and extract year, month, and day as new columns.

df.dtypes
```

```python
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'])
print(df.dtypes)
```

```python
# year, month, and day as new columns
df['Year'] = df['Transaction_Date'].dt.year
df['Month'] = df['Transaction_Date'].dt.month
df['Day'] = df['Transaction_Date'].dt.day

print(df.info())
```

```python
df.head()
df.to_csv("Clean Data.csv")
```

# Data Manipulation and Feature Engineering

* Create a new column Discounted_Amount, assuming a 5% discount on all transactions above #$4500.

* Categorize the Transaction_Amount into Low, Medium, and High based on:
  Low: Below $100
  Medium: Between $100 - $500
  High: Above $500
  
* Drop the Merchant column if more than 30% of values are missing.

```python
#Create a new column Discounted_Amount, assuming a 15% discount on all transactions above $5500.

df['Discounted_Amount'] = df['Amount'].apply(lambda x: x * 0.85 if x > 5500 else x)
df[['Amount', 'Discounted_Amount']].head()
```

```python
#Categorize the Transaction_Amount into Low, Medium, and High based on:
#Low: Below $2000
#Medium: Between $2000 - $4500
#High: Above $4500

df['Amount_Category'] = pd.cut(df['Amount'],
                               bins=[-float('inf'), 2000,4500, float('inf')],
                               labels=['Low', 'Medium', 'High'])
df[['Amount', 'Amount_Category']].head()
```

```python
#Drop the Merchant column if more than 30% of values are missing.

missing_value = df['Merchant'].isnull().mean()
print(missing_value)

if missing_value > 0.3:
    df.drop(columns=['Merchant'], inplace=True)
    print("Deleted missing value...")
else:
    print("No missing value found....")
```

# 3.Aggregation and Insights

* Find the total transaction amount per Category.
* Determine the number of declined transactions per Payment_Mode.
* Identify the top 5 most frequent merchants based on transaction count.

```python
#Find the total transaction amount per Category.

total_amount_per_category = df.groupby('Category')['Amount'].sum().reset_index()
total_amount_per_category
```

```python
#Determine the number of declined transactions per Payment_Mode.

Total_decline_trans = df[df['Transaction_Status'] == 'Declined'].groupby('Payment_Mode').size().reset_index(name='Declined_Count')

Total_decline_trans
```

```python
#Find the average transaction amount per Location.

Ave_Trans_Locationwise = df.groupby('Location')['Amount'].mean().reset_index()
Ave_Trans_Locationwise
```

```python
#Identify the top 5 most frequent merchants based on transaction count (Amount).

Top_Merchant = df['Merchant'].value_counts().head(5).reset_index()
Top_Merchant
```

 # 6️. Data Merging and Joining

* Suppose you have another dataset (customer_info.csv) containing Customer_ID, Age, Gender, and Account_Status.
* Merge it with credit_card_transactions.csv using an appropriate join operation.
* Find the average transaction amount per Age group.

```python
#load customer info CSV.

customer_info = pd.read_csv("customer_info.csv")
customer_info.head()
```

```python
merged_df = pd.merge(df,customer_info, on='Customer_ID', how='inner')
merged_df.head()
```

```python
# Find the average transaction amount per Age group.

age_bin = [0,18,30,50,100]
age_labels = ['<18', '18-30', '31-50', '>50']
merged_df['Age_Group'] = pd.cut(merged_df['Age'],bins=age_bin, labels=age_labels, include_lowest=True)

avg_amount_per_age_group = merged_df.groupby('Age_Group')['Amount'].mean().reset_index()
avg_amount_per_age_group
```
