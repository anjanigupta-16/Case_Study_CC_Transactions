```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("mumbai_west_housing.csv")
df.head()
```

```python
sns.set_style('whitegrid') #clean background
sns.set_palette("viridis") #color
plt.rcParams['figure.figsize'] = (10,6) #size
```

```python
#basic distribution - price spread out 
sns.histplot(data=df, x='Price_Lakhs', kde=True,color='red', bins=35)
plt.title("Distribution - Price Spread Out")
plt.xlabel("Price")
plt.ylabel("Number of House")
plt.show()
```

```python
# BHK most expensive ? Compare Price 

sns.boxplot(data=df,x='BHK', y='Price_Lakhs', palette='Set2', hue='BHK')
plt.title("House Price - Number of BHK")
plt.xlabel("BHK")
plt.ylabel("Price")
plt.show()
```

```python
#scatter plot - Relationship - Price vs Area

sns.scatterplot(data=df,
                x='Area_sqft',
                y='Price_Lakhs',
                hue = 'Locality',
                style='BHK',
                size='BHK',
                palette='tab10'
               )

plt.title("Relationship - Price vs Area")
plt.xlabel("Area Sqft")
plt.ylabel("Price")
plt.legend(bbox_to_anchor=(1.05,1), loc='upper left')
plt.show()
```
